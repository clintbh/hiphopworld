```python
import openpyxl
from geopy.geocoders import ArcGIS

# Load the Excel file and select the sheet
wb = openpyxl.load_workbook('rapdataclean.xlsx')
sheet = wb['Sheet3']

# Create a geolocator object using ArcGIS
geolocator = ArcGIS()

# Iterate through the rows in the sheet, starting with the second row
for i in range(2, sheet.max_row + 1):
    # Get the location from the appropriate column
    location = sheet.cell(row=i, column=3).value
    
    # Use ArcGIS to geocode the location and get its coordinates
    try:
        address, (latitude, longitude) = geolocator.geocode(location)
    except:
        latitude = None
        longitude = None
    
    # Write the coordinates to new columns in the same row
    sheet.cell(row=i, column=4).value = longitude
    sheet.cell(row=i, column=5).value = latitude
    print(location)

# Save the updated Excel file
wb.save('rapdataclean.xlsx')
```
